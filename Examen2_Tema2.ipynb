```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import optuna
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.layers import (
    Dense, Dropout, Conv1D, MaxPooling1D, Flatten,
    LSTM, Bidirectional, TimeDistributed, GlobalAveragePooling1D,
    BatchNormalization, LeakyReLU, GRU, SpatialDropout1D
)

from sklearn.metrics import mean_squared_error, mean_absolute_error
```

### First we load the data

### Then we split it in 70-15-15 splits

```python
# Train set (70%)
train = df.iloc[:int(len(df) * 0.7), :]

# Val set (15%)
val = df.iloc[int(len(df) * 0.7):int(len(df) * 0.85), :]

# Test set (15%)
test = df.iloc[int(len(df) * 0.85):, :]

col = "Total_Orders"

plt.figure(figsize=(12, 6))

plt.plot(train.index, train[col], color="blue", label="Train (70%)")
plt.plot(val.index, val[col], color="orange", label="Val (15%)")
plt.plot(test.index, test[col], color="red", label="Test (15%)")

plt.xlabel("Fecha")
plt.ylabel(col)
plt.legend()
plt.title(f"Target var set split")
plt.show()
```

```python
# Split the multivariate data in steps
def split_multivariate_sequence_m_step(sequence: np.ndarray, n_steps_in: int, n_steps_out: int=1):
    """
    Divide una secuencia multivariada en muestras
    :param sequence: Array de numpy con la secuencia
    :param n_steps_in: Número de pasos de entrada
    :param n_steps_out: Número de pasos de salida
    :return: Devuelve dos arrays de numpy con las muestras de entrada y salida
    """
    X, y = list(), list()
    for i in range(len(sequence)):
        # encontrar el final de este patrón
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out - 1

        # comprobar si estamos más allá de la secuencia
        if out_end_ix > len(sequence):
            break
        # reunir partes de entrada y salida del patrón
        seq_x, seq_y = sequence[i:end_ix, :-1], sequence[end_ix - 1:out_end_ix, -1]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)

n_steps_in, n_steps_out = 6, 1

X_train, y_train = split_multivariate_sequence_m_step(train.values, n_steps_in, n_steps_out)
```

### Now we split in secuences 

```python
#split the data and show shapes
n_steps_in, n_steps_out = 6, 1

X_train, y_train = split_multivariate_sequence_m_step(train.values, n_steps_in, n_steps_out)
X_val, y_val = split_multivariate_sequence_m_step(val.values, n_steps_in, n_steps_out)
X_test, y_test = split_multivariate_sequence_m_step(test.values, n_steps_in, n_steps_out)

print('Train split shape', X_train.shape, y_train.shape)
print('Val split shape', X_val.shape, y_val.shape)
print('Test split shape', X_test.shape, y_test.shape)
```

### Architecture Exploration
We designed **3 architectures for each model**, using a modified version of the pipeline function, now adapted with **deeper networks** to better capture complex patterns in the multivariate data.  
- Each model was trained for **200 epochs**, using **early stopping** and a **fixed learning rate** selected after experimenting with different values.  
- Additionally, a **learning rate scheduler** was applied to reduce the learning rate if no improvement was observed after **20 epochs**, similar to the early stopping strategy and also for 20 epochs.


```python
random_state = 42
np.random.seed(random_state)
tf.random.set_seed(random_state)

# Define Learning rate to experiment
learning_rate = 0.001

# Extract steps and features
n_steps = X_train.shape[1] 
n_features = X_train.shape[2]

###############################################################

# MLP Models

# frist arch from 96 to 24, 4 Dense with relu and one batchnormalization
mlp_model_1 = Sequential([
    Flatten(input_shape=(n_steps, n_features)),
    Dense(96, activation='relu'),
    Dense(48, activation='relu'),
    BatchNormalization(),
    Dense(24, activation='relu'),
    Dense(1)
], name='MLP_Model_1')

# Seccond arch from 160 to 40, 3 Dense 2 Dropout layers with 0.25% 
mlp_model_2 = Sequential([
    Flatten(input_shape=(n_steps, n_features)),
    Dense(160, activation='relu'),
    Dropout(0.25),
    Dense(80, activation='relu'),
    Dropout(0.25),
    Dense(40, activation='relu'),
    Dense(1)
], name='MLP_Model_2')

# Third arch fron 192 to 48, with 3 Dense Leaky relu with alpha 0.1 to allow really small negatives
mlp_model_3 = Sequential([
    Flatten(input_shape=(n_steps, n_features)),
    Dense(192, activation=LeakyReLU(alpha=0.1)),
    BatchNormalization(),
    Dense(96, activation=LeakyReLU(alpha=0.1)),
    Dense(48, activation=LeakyReLU(alpha=0.1)),
    Dense(1)
], name='MLP_Model_3')

# CNN Models

#First arch conv 1d filter with 48 neurons, max pooling by half, and Dense with 64 neurons
cnn_model_1 = Sequential([
    Conv1D(48, 2, activation='relu', padding='same', input_shape=(n_steps, n_features)),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1)
], name='CNN_Model_1')

#Seccond arch, two conv 1D layers with 72 and 144 neurons with leaky relu and max pooling, dense and dropout with 25%
cnn_model_2 = Sequential([
    Conv1D(72, 3, activation=LeakyReLU(alpha=0.1), padding='same', input_shape=(n_steps, n_features)),
    MaxPooling1D(2),
    Conv1D(144, 2, activation=LeakyReLU(alpha=0.1), padding='same'),
    MaxPooling1D(2),
    Flatten(),
    Dense(80, activation=LeakyReLU(alpha=0.1)),
    Dropout(0.25),
    Dense(1)
], name='CNN_Model_2')

# Third arch two conv1d, global average pooling (to get the mean of the secuence) and batchnormmalization
cnn_model_3 = Sequential([
    Conv1D(56, 2, activation='relu', padding='same', input_shape=(n_steps, n_features)),
    BatchNormalization(),
    Conv1D(112, 2, activation='relu', padding='same'),
    GlobalAveragePooling1D(),
    Dense(56, activation='relu'),
    Dense(1)
], name='CNN_Model_3')

# LSTM Models

# First arch
# Stacked Lstm from 40 to 24 neurons with tanh and dropout
lstm_model_1 = Sequential([
    LSTM(40, activation='tanh', return_sequences=True, input_shape=(n_steps, n_features)),
    LSTM(24, activation='tanh'),
    Dropout(0.25),
    Dense(1)
], name='LSTM_Model_1')

# Seccond arch
#Stacked lstm with SpatialDropout (drops entire feature maps instead of only neurons) and a dense layer
lstm_model_2 = Sequential([
    LSTM(72, activation='tanh', return_sequences=True, input_shape=(n_steps, n_features)),
    SpatialDropout1D(0.15),
    LSTM(36, activation='tanh'),
    Dense(16, activation='relu'),
    Dense(1)
], name='LSTM_Model_2')

#Third arch
# Bidirectional GRU with dropout and batchnormalization
lstm_model_3 = Sequential([
    Bidirectional(GRU(56, activation='tanh', return_sequences=True), input_shape=(n_steps, n_features)),
    BatchNormalization(),
    Dense(40, activation='relu'),
    Dropout(0.2),
    Dense(1)
], name='LSTM_Model_3')

# CNN-LSTM

#First arch
# one conv filter and maxpooling, also using Timedistributed dense for the lstm
#Leaky relu cause it has been showing good performance on other models
cnnlstm_model_1 = Sequential([
    Conv1D(72, 2, activation=LeakyReLU(alpha=0.1), padding='same', input_shape=(n_steps, n_features)),
    MaxPooling1D(2),
    TimeDistributed(Dense(30, activation=LeakyReLU(alpha=0.1))),
    LSTM(45, activation='tanh'),
    Dense(1)
], name='CNN_LSTM_Model_1')

#Seccond arch
# two conv filter with normal relu, batch normalization and TimeDistributed for lstm
cnnlstm_model_2 = Sequential([
    Conv1D(56, 2, activation='relu', padding='same', input_shape=(n_steps, n_features)),
    MaxPooling1D(2),
    Conv1D(112, 2, activation='relu', padding='same'),
    BatchNormalization(),
    TimeDistributed(Dense(40, activation='relu')),
    LSTM(50, activation='tanh'),
    Dense(1)
], name='CNN_LSTM_Model_2')

# third arch
# Conv layer, Spatial dropout and maxpooling with timedistributed
cnnlstm_model_3 = Sequential([
    Conv1D(80, 3, activation=LeakyReLU(alpha=0.1), padding='same', input_shape=(n_steps, n_features)),
    SpatialDropout1D(0.1),
    MaxPooling1D(2),
    TimeDistributed(Dense(36, activation=LeakyReLU(alpha=0.1))),
    Bidirectional(GRU(48, activation='tanh')),
    Dense(1)
], name='CNN_LSTM_Model_3')

# get the models for training
models = [
    mlp_model_1, mlp_model_2, mlp_model_3,
    cnn_model_1, cnn_model_2, cnn_model_3,
    lstm_model_1, lstm_model_2, lstm_model_3,
    cnnlstm_model_1, cnnlstm_model_2, cnnlstm_model_3
]

#this learning rate schedule reduces lr after 20 epochs of not improoving
lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=20,
    min_lr=1e-6,
    verbose=1
)

#Early stopping as in the univariate example, 20 epochs as patience
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=20,
    restore_best_weights=True,
    verbose=1
)

# compile all the models
for model in models:
    model.compile(
        optimizer=Adam(learning_rate=learning_rate, clipnorm=1.0),
        loss='mse',
        metrics=['mae']
    )

# Train and evaluate each model
def train_and_evaluate_model(model, X_train, y_train, X_val, y_val, X_test, y_test, epochs=200, batch_size=64):
    print(f'\nTraining {model.name}...')
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[lr_scheduler, early_stopping],
        verbose=1
    )
    
    # Eval on the test set
    test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
    print(f'{model.name} - Test Loss: {test_loss:.4f}, Test MAE: {test_mae:.4f}')
    
    return history, test_loss, test_mae

# train and eval each model in the list
results = {}
for model in models:
    history, test_loss, test_mae = train_and_evaluate_model(
        model, X_train, y_train, X_val, y_val, X_test, y_test
    )
    results[model.name] = {
        'history': history,
        'test_loss': test_loss,
        'test_mae': test_mae
    }

# Print results
print("\n===== MODEL PERFORMANCE SUMMARY =====")
for model_name in results:
    print(f"{model_name}: Test MAE = {results[model_name]['test_mae']:.4f}")

# # Find the best model based on the test MAE
best_model_name = min(results, key=lambda x: results[x]['test_mae'])
print(f"\nBest model: {best_model_name} with Test MAE = {results[best_model_name]['test_mae']:.4f}")
```

The best model was a **MLP** with a Test MAE 66.9499:  
- **1 Flatten layer** for input shape
- **1 Dense layer** with **192 neurons** and **LeakyReLU activation (alpha=0.1)**  
- **1 BatchNormalization layer**  
- **1 Dense layer** with **96 neurons** and **LeakyReLU activation (alpha=0.1)**  
- **1 Dense layer** with **48 neurons** and **LeakyReLU activation (alpha=0.1)**  
- **1 Dense output layer** with **1 neuron**  

It outperformed by far the other models in this experiments, also after some redefining of the architectures


### Final Evaluation
Now we're going to take the **best model architecture** and use it to **predict on the test set** to evaluate how well it generalizes.  
- The test set contains only **4 samples**, due to the **sequence window size** we selected.  
- Even though it's a small sample, we ensured that these **4 points were never seen by the model** during training, making it a valid test for generalization performance.  
- It will also be used as a **comparison metric against the Optuna-optimized model** to better assess relative performance.


```python
# get the best model from the previous experiment
best_model_name = min(results, key=lambda x: results[x]['test_mae'])
best_model = next(model for model in models if model.name == best_model_name)

# Generate predictions on the test split
y_pred = best_model.predict(X_test)

# Convert predictions and actual values to 1D arrays if needed
y_test_flat = y_test.flatten() if y_test.ndim > 1 else y_test
y_pred_flat = y_pred.flatten() if y_pred.ndim > 1 else y_pred

# Calculate metrics
mse = mean_squared_error(y_test_flat, y_pred_flat)
mae = mean_absolute_error(y_test_flat, y_pred_flat)
rmse = np.sqrt(mse)

# Create sequence for the x axis
x = np.arange(len(y_test_flat))

# Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(x, y_test_flat, label='Actual Values', color='blue', linewidth=2)
plt.plot(x, y_pred_flat, label='Predicted Values', color='red', linewidth=1.5, alpha=0.8)

# Scatterplot
plt.scatter(x, y_test_flat, color='blue', s=10, alpha=0.5)
plt.scatter(x, y_pred_flat, color='red', s=10, alpha=0.5)

# Highlight discrepancies with vertical lines
for i in x:
    plt.plot([i, i], [y_test_flat[i], y_pred_flat[i]], 'gray', alpha=0.3, linestyle='--')

# Add labels and title
plt.title(f'Predictions vs Actual Values using {best_model_name}', fontsize=16)
plt.xlabel('Test Sample Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Add metrics
plt.figtext(0.15, 0.8, f'RMSE: {rmse:.4f}\nMAE: {mae:.4f}', 
            bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.5'))

plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

print(f"\nSample of Predictions vs Actual Values for {best_model_name}:")
print("-" * 60)
print(f"{'Index':>8} | {'Actual':>12} | {'Predicted':>12} | {'Difference':>12}")
print("-" * 60)

n_samples = min(20, len(y_test_flat))
for i in range(n_samples):
    actual = y_test_flat[i]
    pred = y_pred_flat[i]
    diff = actual - pred
    print(f"{i:8d} | {actual:12.4f} | {pred:12.4f} | {diff:12.4f}")

plt.tight_layout()
plt.show()
```

# Optuna optimization

After that, we select the model that gave the best results and optimize it with **Optuna** for:  
- **Learning rate**  
- **Activation functions**  
- **Number of layers**  
- **Number of units per layer**  
- **Optimizer**  

The function is designed to **adapt** and tune the best model architecture **regardless of its type**, as different models can perform better depending on the series. Our goal is to achieve the **optimal performance** possible.  

The experiment runs for **50 trials**, then the results are **de-scaled** and plotted against the **test set**, allowing us to compare with the previous model.  

```python
# Define the objective function, and give parameters to optimize layers, neurons, lr, optimizer and activation functions
def objective(trial):
    n_layers = trial.suggest_int('n_layers', 1, 5)
    units = [trial.suggest_int(f'units_{i}', 32, 256) for i in range(n_layers)]
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-1)
    optimizer_name = trial.suggest_categorical('optimizer', ['adam', 'sgd'])
    activation_function = trial.suggest_categorical('activation_function', ['relu', 'tanh', 'sigmoid'])

    model = Sequential()
    model.add(Flatten(input_shape=(n_steps, n_features)))

    for unit in units:
        model.add(Dense(unit, activation=activation_function))
        model.add(BatchNormalization())

    model.add(Dense(1))

    optimizer = Adam(learning_rate=learning_rate) if optimizer_name == 'adam' else SGD(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])

    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=200, batch_size=64, callbacks=[early_stopping], verbose=0)

    test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
    return test_mae

# Create a study and optimize for 50 trials
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=50)

# get the best trial
best_trial = study.best_trial
print(f'Best trial: {best_trial.number}')
print(f'Best MAE: {best_trial.value}')
print(f'Best hyperparameters: {best_trial.params}')

# Train the model with the hyperparams of the best trial
best_params = best_trial.params
n_layers = best_params['n_layers']
units = [best_params[f'units_{i}'] for i in range(n_layers)]
learning_rate = best_params['learning_rate']
optimizer_name = best_params['optimizer']
activation_function = best_params['activation_function']

# build and train the best model
best_model = Sequential()
best_model.add(Flatten(input_shape=(n_steps, n_features)))

for unit in units:
    best_model.add(Dense(unit, activation=activation_function))
    best_model.add(BatchNormalization())

best_model.add(Dense(1))

optimizer = Adam(learning_rate=learning_rate) if optimizer_name == 'adam' else SGD(learning_rate=learning_rate)
best_model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])
best_model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=200, batch_size=64, verbose=1)

# Generate predictions
y_pred = best_model.predict(X_test)

# Flatten the arrays if needed
y_test_flat = y_test.flatten() if y_test.ndim > 1 else y_test
y_pred_flat = y_pred.flatten() if y_pred.ndim > 1 else y_pred

# Evaluate
mse = mean_squared_error(y_test_flat, y_pred_flat)
mae = mean_absolute_error(y_test_flat, y_pred_flat)
rmse = np.sqrt(mse)

# Plot the same graph than the previous experiment
plt.figure(figsize=(12, 6))
x = np.arange(len(y_test_flat))
plt.plot(x, y_test_flat, label='Actual Values', color='blue', linewidth=2)
plt.plot(x, y_pred_flat, label='Predicted Values', color='red', linewidth=1.5, alpha=0.8)
plt.scatter(x, y_test_flat, color='blue', s=10, alpha=0.5)
plt.scatter(x, y_pred_flat, color='red', s=10, alpha=0.5)

for i in x:
    plt.plot([i, i], [y_test_flat[i], y_pred_flat[i]], 'gray', alpha=0.3, linestyle='--')

plt.title(f'Predictions vs Actual Values using Best Model', fontsize=16)
plt.xlabel('Test Sample Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.figtext(0.15, 0.8, f'RMSE: {rmse:.4f}\nMAE: {mae:.4f}', bbox=dict(facecolor='white', alpha=0.8, boxstyle='round,pad=0.5'))
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# predictions vs actual values
print(f"\nSample of Predictions vs Actual Values:")
print("-" * 60)
print(f"{'Index':>8} | {'Actual':>12} | {'Predicted':>12} | {'Difference':>12}")
print("-" * 60)

n_samples = min(20, len(y_test_flat))
for i in range(n_samples):
    actual = y_test_flat[i]
    pred = y_pred_flat[i]
    diff = actual - pred
    print(f"{i:8d} | {actual:12.4f} | {pred:12.4f} | {diff:12.4f}")
```
