```python
import pandas as pd
from pipeline import (univariate_data_analysis, load_data, split_data, optimize_model, 
                       time_series_forecasting, plot_model_predictions)
def index_date(ts):
    starting_date = df_info_filtered[df_info_filtered['M4id'] == ts]['StartingDate'].iloc[0]
    date_range = pd.date_range(start=starting_date, periods=len(df_filtered[ts]), freq='H')
    df_filtered[ts].index = date_range
    return df_filtered[ts].dropna()

# Load and prepare data
df_filtered, df_info_filtered = load_data()
ts = index_date('H38')
univariate_data_analysis(ts)
train_data, val_data, test_data = split_data(ts)

# Get predictions from time series forecasting
standard_model_result = time_series_forecasting(train_data, val_data, test_data)
standard_model = standard_model_result['best_model']
standard_forecast = standard_model_result['forecast']

# Get predictions from optimized model
optimized_model_result = optimize_model(train_data, val_data, test_data, n_steps=24, model_type='mlp', n_trials=2)
optimized_model = optimized_model_result['best_model']
optimized_forecast = optimized_model_result['forecast']

# Prepare models info for plotting
models_info = [
    (standard_model, 'Standard Model', standard_forecast),
    (optimized_model, 'Optimized Model', optimized_forecast)
]

# Plot predictions
plot_model_predictions(models_info, train_data, val_data, test_data, forecast_horizon=48)
```
